```python
from scipy.spatial.distance import cityblock
```

```python
distances = []
```

```python
def Recommendation(track_name, N):
    if track_name not in df['track_name'].values:
        # If the track is not found, recommend N songs by combined popularity
        top_recommendations = df.sort_values(by='combined_popularity', ascending=False).head(N)
        return top_recommendations[['track_name']]

    target_song = df[df['track_name'] == track_name].iloc[0, 2:]  # Exclude non-numeric columns

    #  Manhattan Distance to all songs
    distances = df.iloc[:, 2:].apply(lambda x: cityblock(target_song, x), axis=1)

    # DataFrame with distances and track names
    similar_songs = pd.DataFrame({'track_name': df['track_name'], 'distance': distances})

    # Sort by distance and exclude the target song
    similar_songs = similar_songs.sort_values(by='distance', ascending=True)
    similar_songs = similar_songs[similar_songs['track_name'] != track_name]

    top_recommendations = similar_songs.head(N)

    return top_recommendations[['track_name']]
```

```python
input_track = 'Orange'
N = 10
recommended_songs = Recommendation(input_track, N)
print(f"Recommendations for '{input_track}':")
```

```python
import pandas as pd

# Assume 'rec' is your DataFrame with preprocessed data

def recommend_songs(input_track_name, num_recommendations, weights):
    # Find the artist name of the input track
    input_track_data = rec[rec['track_name'] == input_track_name]
    input_artist_name = input_track_data['artist_name'].values[0]
    input_genres = input_track_data.iloc[:, 3:].columns[input_track_data.iloc[:, 3:].values[0] > 0].tolist()

    # Find songs by the same artist
    same_artist_songs = rec[rec['artist_name'] == input_artist_name]

    # Find songs with matching genres
    genre_mask = rec[rec.columns[3:]].apply(lambda row: any(genre in input_genres for genre in row.index[row > 0]), axis=1)
    matching_genre_songs = rec[genre_mask]

    # Combine recommendations with weights
    combined_recommendations = {
        'Same Artist': same_artist_songs['track_name'].tolist(),
        'Matching Genres': matching_genre_songs['track_name'].tolist(),
    }

    weighted_recommendations = {}
    for method, recommended_tracks in combined_recommendations.items():
        weighted_recommendations[method] = recommended_tracks[:num_recommendations]

    # Sort the recommendations based on the weights
    sorted_recommendations = sorted(
        weighted_recommendations.items(),
        key=lambda item: weights[item[0]],
        reverse=True
    )

    # Extract the final recommended tracks
    final_recommendations = []
    for method, recommended_tracks in sorted_recommendations:
        final_recommendations.extend(recommended_tracks)

    # Remove duplicates and return the top num_recommendations
    final_recommendations = list(set(final_recommendations))
    return final_recommendations[:num_recommendations]

# Example input
input_track_name = 'The Real Slim Shady'  # Replace with your desired input track
num_recommendations = 10  # Number of songs to recommend
weights = {
    'Same Artist': 0.6,
    'Matching Genres': 0.4,
}

# Get recommended songs
recommended_songs = recommend_songs(input_track_name, num_recommendations, weights)

# Display the recommended songs
print("Recommended Songs:")
for song in recommended_songs:
    print(song)
```
